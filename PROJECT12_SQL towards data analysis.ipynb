Oftentimes we will want to enrich our data by adding more data from other datasets to it.
There are two major ways we can combine our datasets together.

# Union

a union (SQL union) is the process of stacking two tables on top of one another. You will usually do this when your data is split up into multiple sections like an excel spreadsheet of a year sales split by month
